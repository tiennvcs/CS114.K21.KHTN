## Lý do chọn đề tài

- Giả mạo khuôn mặt là một phương pháp ứng dụng trí tuệ nhân tạo nhằm mạo danh danh tính của người khác.

- Xét về phương diện học thuật, giả mạo khuôn mặt (Fake face) là một thành tựu của Trí tuệ nhân tạo nên bài toán này chính là dùng Trí tuệ nhân tạo để CHỐNG lại Trí tuệ nhân tạo.
